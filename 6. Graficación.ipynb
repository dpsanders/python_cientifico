## Enfoque más orientado a objetos de matplotlib

Más control sobre la gráfica al interactuar a un nivel más bajo, por ejemplo dos subgráficas en una sola gráfica:

En matplotlib, hay *figura* y *axes* ("ejes"): puede haber varios ejes en una sola figura. La figura es el conjunto de todo

Por ejemplo, para tener 2 subfiguras:

```python
fig, ax = plt.subplots(ncols=2)

print ax[0], ax[1]
```

```python
fig, ax = plt.subplots(ncols=2)

ax[0].plot(x, y)
ax[1].plot(x, x**3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[1, 0].plot(x, y); # matriz de plots
```

Cuadrado unitario como polígono relleno:

```python
x = np.r_[0, 1, 1, 0]
y = np.r_[0, 0, 1, 1]

plt.fill(x, y, alpha=0.5)

# plt.xlim(-0.1, 2.1)
# plt.ylim(-0.1, 2.1)

plt.fill(x+0.5, y+0.5, alpha=0.5, edgecolor="red", color="magenta")

plt.axis("scaled")
```

Ejercicio: Dibujen así un círculo

O pueden utilizar un *patch* (*parche*):

```python
from matplotlib import patches
```

```python
c = patches.Circle( (1, 1), 0.5 )
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.add_patch(c)

plt.axis("scaled")

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()
```

## Datos 3-dimensionales y sus representaciones

Lo pienso como una discretización de una función $\mathbb{R}^2 \to \mathbb{R}$

Hay varias maneras de dibujarla:

```python
plt.contour(M)
plt.axis("scaled")
```

```python
plt.contourf(M)
```

```python
plt.pcolor(M)
```

```python
plt.imshow(M)
```

```python
x = np.linspace(-2, 2, 5)
y = np.linspace(-2, 2, 5)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
X
```

```python
Y
```

```python
Z = X**2 + Y**2
```

```python
plt.contourf(X, Y, Z)
```

```python
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)

X, Y = np.meshgrid(x, y)
Z = X*X + Y*Y

plt.contourf(X, Y, Z)
plt.axis("scaled")
```

```python
plt.pcolormesh(X, Y, Z, shading="gouraud")
plt.axis("scaled")
```

## Campos vectoriales

```python
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)

X, Y = np.meshgrid(x, y)
plt.quiver(X, Y, Y, -X, pivot="middle")   # carcaj: donde se guardan las flechas...\
plt.streamplot(X, Y, Y, -X)
plt.axis("scaled")
```

```python
plt.streamplot(X, Y, Y, -X)
plt.axis("scaled")
```
